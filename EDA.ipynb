```python
import os

# define directory paths
fresh_dir = 'data/freshoranges/'
rotten_dir = 'data/rottenoranges/'
fresh_images = os.listdir(fresh_dir)
rotten_images = os.listdir(rotten_dir)

print(f'Number of fresh oranges: {len(fresh_images)}')
print(f'Number of rotten oranges: {len(rotten_images)}')
```

```python
import matplotlib.pyplot as plt
import random

def display_sample_images(image_paths, title):
    plt.figure(figsize=(10, 10))
    if title == "Fresh Oranges":
        path = fresh_dir
    else:
        path = rotten_dir
    for i in range(9):
        plt.subplot(3, 3, i+1)
        img = plt.imread(os.path.join(path, random.choice(image_paths)))
        plt.imshow(img)
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

display_sample_images(fresh_images, 'Fresh Oranges')
display_sample_images(rotten_images, 'Rotten Oranges')
```

```python
import numpy as np
def analyze_image_properties(image_paths, title):
    sizes = []
    aspect_ratios = []
    if title == "fresh":
        path = fresh_dir
    else:
        path = rotten_dir
    for img_path in image_paths:
        img = plt.imread(os.path.join(path, img_path))
        sizes.append(img.shape[:2])
        aspect_ratios.append(float(img.shape[1]) / float(img.shape[0]))

    sizes = np.array(sizes)
    aspect_ratios = np.array(aspect_ratios)

    print('Size (min, max):', sizes.min(axis=0), sizes.max(axis=0))
    print('Aspect ratio (min, max):', aspect_ratios.min(), aspect_ratios.max())

analyze_image_properties(fresh_images, "fresh")
analyze_image_properties(rotten_images, "rotten")
```
